```python
import random
import numpy as np
import math as mt
m=random.uniform(-10,10)
c=random.uniform(-10,10)
eta=0.9
ssg_m=0
ssg_c=0
max_iter=10000
eps=mt.pow(10,-8)
data=np.array([[0.2,3.4],[0.4,3.8],[0.6,4.2],[0.8,4.6]])
```

```python
def gradient(m_f,c_f,x,y):
  m_deriv=-1*(y-m_f*x-c_f)*x
  c_deriv=-1*(y-m_f*x-c_f)
  return m_deriv,c_deriv
```

```python
for iter in range(max_iter):
  del_m,del_c=0.0,0.0
  for sample in range(len(data[:0])):
    delm,delc=gradient(m,c,data[sample,0],data[sample,1])
    del_m=del_m+delm
    del_c=del_c+delc
  del_m=del_m/len(data[:,0])
  del_c=del_c/len(data[:,0])
  ssg_m=ssg_m+(del_m*del_m)
  ssg_c=ssg_c+(del_c*del_c)
  m=m-(eta*del_m)/(mt.sqrt(ssg_m+eps))
  c=c-(eta*del_c)/(mt.sqrt(ssg_c+eps))

print(f"\nThe local minima occurs at: {m}, {c}")
```

```python
normalized_df=(df-df.min())/(df.max()-df.min())
normalized_df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(normalized_df.Day_1,normalized_df.Day_2,test_size=0.10,random_state=42)
```

```python
import random
import numpy as np
import math as mt
m=random.uniform(-10,10)
c=random.uniform(-10,10)
eta=0.9
ssg_m=0
ssg_c=0
max_iter=10000
eps=mt.pow(10,-8)
```

```python
def gradient(m_f,c_f,x,y):
  m_deriv=-1*(y-m_f*x-c_f)*x
  c_deriv=-1*(y-m_f*x-c_f)
  return m_deriv,c_deriv
```

```python
for iter in range(max_iter):
  del_m,del_c=0.0,0.0
  for sample in range(x_train.shape[0]):
    delm,delc=gradient(m,c,x_train.iloc[sample],y_train.iloc[sample])
    del_m=del_m+delm
    del_c=del_c+delc
  del_m=del_m/x_train.shape[0]
  del_c=del_c/x_train.shape[0]
  ssg_m=ssg_m+(del_m*del_m)
  ssg_c=ssg_c+(del_c*del_c)
  m=m-(eta*del_m)/(mt.sqrt(ssg_m+eps))
  c=c-(eta*del_c)/(mt.sqrt(ssg_c+eps))

print(f"\nThe local minima occurs at: {m}, {c}")
```

```python
import numpy as np
x_train=np.array(x_train)
y_train=np.array(y_train)
x_test=np.array(x_test)
y_test=np.array(y_test)
```

```python
y_pred_test=[]
for i in x_test:
  y_p_te=(m*i)+c
  y_pred_test.append(y_p_te)
y_pred_test=np.array(y_pred_test)
```

```python
import math
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

rmse=math.sqrt(mean_squared_error(y_test,y_pred_test))
print('Root Mean Squared Error:',rmse)
mse=mean_squared_error(y_test,y_pred_test)
print('Mean Squared Error:',mse)
mae=mean_absolute_error(y_test,y_pred_test)
print('Mean Absolute Error:',mae)
```
